```python
#Building a Decision Tree classification
#Load tge Dataset
import pandas as pd
bank_df = pd.read_csv("bank_kaggle.csv")
bank_df.head()
```

```python
#WOrk with X(Predictor Variables)
X_features = list(bank_df.columns)
X_features.remove('deposit')
X_features
```

```python
#Encode the Categorical Variables
encode_bank_df = pd.get_dummies(bank_df[X_features])
```

```python
encode_bank_df.head()
```

```python
#Set the X and Y
X = encode_bank_df
Y = bank_df.deposit.map(lambda x: int(x=='yes'))
```

```python
#split the Dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=25)
```

```python
#Build the decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

clf_tree = DecisionTreeClassifier(criterion = 'gini', max_depth= 3)
```

```python
clf_tree.fit(X_train,Y_train)
```

```python
#Accuracy Score
clf_tree.score(X_test, Y_test)
```
