```python
#4.Display the Month with most of loans issued date

data['month_dt'] = pd.to_datetime(data['date_issued'])

data['month']= data['month_dt'].dt.month
data.drop('month_dt',axis =1)
print ("The month with the most of loans issued date = ",data['month'].max())
```
