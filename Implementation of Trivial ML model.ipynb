```python
#Accuracy 
def get_predictions(X_binarised_train, Y_train,b):
    pred=[]
    for i in range(X_binarised_train.shape[0]):
        if (np.sum(X_binarised_train.shape[0])):
            pred.append(1)
        else:
            pred.append(0)
    df=pd.DataFrame(columns=['predictions','True_values'])
    df['predictions']=pred
    df['True_values']=Y_train
    df['cor_wr']=df['predictions']==df['True_values']
    display(df['cor_wr'].value_counts())
    return df['cor_wr'].value_counts()[1] / (df['cor_wr'].value_counts())
```

```python
#Calculating Accuracy
no=X_train.shape[1]
for b in range(no):
    print('value of b:', b)
    x=get_predictions(X_binarised_train, Y_train, b)
    print('accuracy: ',x)
```
